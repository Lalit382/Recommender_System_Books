```python
#Importing Necessary Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Reading the Files
books=pd.read_csv("Books.csv")
Users=pd.read_csv("Users.csv")
Ratings=pd.read_csv("Ratings.csv")
```

# Popularity Based Recommender System

```python
#Checking the files for any sort of duplicated values
print(books.duplicated().sum())
print(Ratings.duplicated().sum())
print(Users.duplicated().sum())
```

```python
#Merging the Ratings and books file 
ratings_name=Ratings.merge(books,on="ISBN")
```

```python
#Selecting only the Book_Title and the rating counts
num_rating=ratings_name.groupby("Book-Title").count()["Book-Rating"].reset_index()
num_rating.rename(columns={"Book-Rating":"No_Of-Ratings"},inplace=True)
num_rating
```

```python
#Calculations of the average _ratings
avg_rat=ratings_name.groupby("Book-Title").mean()["Book-Rating"].reset_index()
avg_rat.rename(columns={"Book-Rating":"Average_Ratings"},inplace=True)
avg_rat
```

```python
#Merging the count and average ratings dataframes on book title
ui_mat=num_rating.merge(avg_rat,on="Book-Title")
new_mat=ui_mat[ui_mat["No_Of-Ratings"]>=250].sort_values("Average_Ratings",ascending=False)
popular_df=new_mat.head(50)
```

```python
#Formation of popular dataframe by merging the above with books file
popular_df=popular_df.merge(books,on="Book-Title").drop_duplicates("Book-Title")[["Book-Title","No_Of-Ratings","Average_Ratings","Image-URL-M","Year-Of-Publication","Book-Author"]]
```

# Collaborative Based Recommender System 

```python
#Selecting the users who have rated more than 200 movies
x=ratings_name.groupby("User-ID").count()["Book-Rating"]>200
frequent_raters=x[x].index
filtered_ratings=ratings_name[ratings_name["User-ID"].isin(frequent_raters)]
filtered_ratings
```

```python
#Selecting the movie which have been rated atleast 50 ratings
y=filtered_ratings.groupby("Book-Title").count()['Book-Rating']>=50
famous_books=y[y].index
final_ratings=filtered_ratings[filtered_ratings["Book-Title"].isin(famous_books)]
```

```python
final_ratings
```

```python
#FORMATION OF USER INTERACTION MATRIX
UI_MAT=final_ratings.pivot_table(index="Book-Title",columns="User-ID",values="Book-Rating")
UI_MAT
```

```python
#Filling the Nan values with 0 
UI_MAT.fillna(0,inplace=True)
```

```python
#Distance metric used are cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#Similarity scores with one movie with the other movies
similarity_scores=cosine_similarity(UI_MAT)
similarity_scores.shape
```

```python
#Defining the Recommendation system
def recommend(book_name):
    #fetching Index
    index=np.where(UI_MAT.index==book_name)[0][0]
    similar_items=sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]
    for i in similar_items:
        print(UI_MAT.index[i[0]])
```
